```python
import numpy as np
from matplotlib import pyplot
from ipywidgets import interact
import sys
from os import path

avg_dir='kw_avg_data/'
out_dir='kw_data_files_dis/'

Wsc=120
W=Wsc+40
# Lsc_sw=np.arange(10,50,10)
Lsc_sw=np.array([10])

# parameters
mu=0.6
mu_sc=1.5
phi=0.05
U0=0.5 # disorder strength
Delta=0.06
t_j=0.6

def wf_avg(SC_width):
    Lsc=SC_width
    L=Lsc+40 # total length of system
    Nrep=400
    N_avg=0
    pyplot.figure()
    for i_r in range(Nrep):
        fname=out_dir+ 'hj_phi_%.2f_mu_%.2f_mus_%.2f_D_%.2f_tj_%.2f_W_%d_L_%d_Ws_%d_Ls_%d_r_%d.npz' % (phi,mu,mu_sc,Delta,t_j,W,L,Wsc,Lsc,i_r)
        if path.exists(fname):
            N_avg += 1
            npzfile = np.load(fname)
            if N_avg==1:
                E_list=npzfile['E_list']
                wf_arr=npzfile['wf_arr']
                Pe=npzfile['Pe']
                Ph=npzfile['Ph']
            else:
                pyplot.plot(E_list/Delta,npzfile['Pe']-npzfile['Ph'])
                wf_arr += npzfile['wf_arr']
                Pe += npzfile['Pe']
                Ph += npzfile['Ph']
    pyplot.show()
    if N_avg>0:
        wf_arr= wf_arr/N_avg
        Pe= Pe/N_avg
        Ph= Ph/N_avg    
        fname=avg_dir+ 'hj_avg_phi_%.2f_mu_%.2f_mus_%.2f_D_%.2f_tj_%.2f_W_%d_L_%d_Ws_%d_Ls_%d.npz' % (phi,mu,mu_sc,Delta,t_j,W,L,Wsc,Lsc)
        np.savez(fname, E_list=E_list, Pe=Pe, Ph=Ph, wf_arr=wf_arr)
        
    return N_avg

for i_L in range(len(Lsc_sw)):
    N_avg= wf_avg(Lsc_sw[i_L])
    print("Average is computed over %d reps." % (N_avg) )
```
