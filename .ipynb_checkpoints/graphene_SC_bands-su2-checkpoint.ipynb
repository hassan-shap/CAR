```python
import kwant
import numpy as np
from cmath import exp
from math import pi, sqrt, tanh
# For plotting
# %run matplotlib_setup.ipy
import matplotlib.pyplot as plt
from ipywidgets import interact
import time
import sys
import tinyarray
from kwant.digest import uniform    # a (deterministic) pseudorandom number generator
```

```python
# armchair edges

t=1.0
t_sc=0.5

# Define the graphene lattice

tau_x = tinyarray.array([[0, 1], [1, 0]])
tau_y = tinyarray.array([[0, -1j], [1j, 0]])
tau_z = tinyarray.array([[1, 0], [0, -1]])


def make_system(Delta=0.2,
                W=200, L=200, Lsc=20, t_j=0.1, mu=0.6, mu_sc=2, phi=0):

            
    def qh_slab(pos):
        (x, y) = pos 
        return (0 <= x < W) and (Lsc/2 <= abs(y) < L+Lsc/2)

    def hopping_ab(site_i, site_j, phi):
        xi, yi = site_i.pos
        xj, yj = site_j.pos
#         # modulated hopping in x direction
        H1=tinyarray.array([[-t*exp(-1j * pi* phi * (xi - xj) * (yi + yj-Lsc*np.sign(yi + yj))),0],\
                            [0,t*exp(1j * pi* phi * (xi - xj) * (yi + yj-Lsc*np.sign(yi + yj)))]])
        return H1

    def onsite(site, mu):
        return  - mu*tau_z

    
    # Define the graphene lattice
    sin_30, cos_30 = (1 / 2, sqrt(3) / 2)
    lat = kwant.lattice.general([(sqrt(3), 0), (0, 1)],
                                [(0, 0), (1/sqrt(3), 0), (3/sqrt(3)/2, 1/2),(5/sqrt(3)/2, 1/2)],norbs=2)
    subA1, subB1,subA2, subB2 = lat.sublattices
    syst = kwant.Builder()
    syst[lat.shape(qh_slab, (0,Lsc))] = onsite
    syst[lat.shape(qh_slab, (0,-Lsc))] = onsite
    syst[lat.neighbors()] = hopping_ab

    def hopping_jn(site_i, site_j, t_j):
        return -t_j*tau_z

    # sc part
    def onsite_sc(site, mu_sc, Delta):
        return  (- mu_sc + 4 * t_sc)* tau_z + Delta *tau_x

    def hopping_sc(site1, site2):
        return -t_sc* tau_z
    
    a0=3/(4*sqrt(3))
    b0=0.5
    primitive_vectors = [(a0, 0), (0, b0)]
    lat_sc = kwant.lattice.Monatomic(primitive_vectors, offset=(a0/6,0),norbs=2)

    Wsq=int(W/a0)
    Lsq=int(Lsc/a0)
    syst[(lat_sc(x,y) for x in range(Wsq) for y in range(-int(Lsq/2)+1,int(Lsq/2)))] = onsite_sc
    syst[((lat_sc(4*i,int(Lsq/2)-1), subA1(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
    syst[((lat_sc(4*i+1,int(Lsq/2)-1), subB1(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
    syst[((lat_sc(4*i+2,int(Lsq/2)-1), subA2(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
    syst[((lat_sc(4*i+3,int(Lsq/2)-1), subB2(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
    syst[((lat_sc(4*i,-int(Lsq/2)+1), subA1(i,-int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
    syst[((lat_sc(4*i+1,-int(Lsq/2)+1), subB1(i,-int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
    syst[((lat_sc(4*i+2,-int(Lsq/2)+1), subA2(i,-int(Lsc/2)-1)) for i in range(int(Wsq/4)))] = hopping_jn
    syst[((lat_sc(4*i+3,-int(Lsq/2)+1), subB2(i,-int(Lsc/2)-1)) for i in range(int(Wsq/4)))] = hopping_jn


    sym_right = kwant.TranslationalSymmetry(lat.vec((1, 0)))
    right_lead = kwant.Builder(sym_right)
    right_lead[lat.shape(qh_slab, (0,Lsc))] = onsite
    right_lead[lat.shape(qh_slab, (0,-Lsc))] = onsite
    right_lead[lat.neighbors()] = hopping_ab
    
    right_lead[(lat_sc(x,y) for x in range(4) for y in range(-int(Lsq/2)+1,int(Lsq/2)))] = onsite_sc
    right_lead[kwant.builder.HoppingKind((0,1),lat_sc,lat_sc)] = hopping_sc
    right_lead[kwant.builder.HoppingKind((1,0),lat_sc,lat_sc)] = hopping_sc
    right_lead[((lat_sc(4*i,int(Lsq/2)-1), subA1(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
#     right_lead[((lat_sc(4*i+1,int(Lsq/2)-1), subB1(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
#     right_lead[((lat_sc(4*i+2,int(Lsq/2)-1), subA2(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
#     right_lead[((lat_sc(4*i+3,int(Lsq/2)-1), subB2(i,int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
#     right_lead[((lat_sc(4*i,-int(Lsq/2)+1), subA1(i,-int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
#     right_lead[((lat_sc(4*i+1,-int(Lsq/2)+1), subB1(i,-int(Lsc/2))) for i in range(int(Wsq/4)))] = hopping_jn
#     right_lead[((lat_sc(4*i+2,-int(Lsq/2)+1), subA2(i,-int(Lsc/2)-1)) for i in range(int(Wsq/4)))] = hopping_jn
#     right_lead[((lat_sc(4*i+3,-int(Lsq/2)+1), subB2(i,-int(Lsc/2)-1)) for i in range(int(Wsq/4)))] = hopping_jn


    right_lead[(lat_sc(0,int(Lsq/2)-1), subA1(0,int(Lsc/2)))] = hopping_jn
    right_lead[(lat_sc(1,int(Lsq/2)-1), subB1(0,int(Lsc/2)))] = hopping_jn
    right_lead[(lat_sc(2,int(Lsq/2)-1), subA2(0,int(Lsc/2)))] = hopping_jn
    right_lead[(lat_sc(3,int(Lsq/2)-1), subB2(0,int(Lsc/2)))] = hopping_jn
    right_lead[(lat_sc(0,-int(Lsq/2)+1), subA1(0,W-1))] = hopping_jn
    right_lead[(lat_sc(1,-int(Lsq/2)+1), subB1(0,W-1))] = hopping_jn
    right_lead[(lat_sc(2,-int(Lsq/2)+1), subA2(0,W-1))] = hopping_jn
    right_lead[(lat_sc(3,-int(Lsq/2)+1), subB2(0,W-1))] = hopping_jn


    syst.attach_lead(right_lead)
    syst.attach_lead(right_lead.reversed())

    
    return syst
```

```python
# Wsc must be multiples of eight
W=6
L=4
Lsc=4
syst = make_system(W=W, L=L, Lsc=Lsc)
# Check that the system looks as intended.
# kwant.plot(syst, colorbar=False,fig_size=(40,80),num_lead_cells=2)
kwant.plot(syst,num_lead_cells=2)
# Finalize the system.
syst = syst.finalized()
```

```python
mu=0.25
mu_sc= mu+ t
Delta=0.03
t_j=.0
gs=0.
gn=0.
lam=0.0
phi=0.004

momenta = [(-pi + 0.01 * pi * i) for i in range(201)]
# def plot_wf(phi=0.03):
# params=dict(t_j=t_j, gn=gn, lam=lam, gs=gs, Delta=Delta, mu=mu ,mu_sc=mu_sc,phi=phi,momenta=momenta)
# fig = plt.figure()  # create a figure object
# ax1 = fig.add_subplot(1, 1, 1)  # create an axes object in the figure
# kwant.plotter.bands(syst.leads[0],params=params,ax=ax1)
# ax1.set_ylim(-0.06,0.06)
# #     ax1.set_xlim(-3,-1)
# plt.show()
    
# interact(plot_wf, phi=(0, 0.1, 0.002));
# plot_wf(phi=0.0095*4.5)

params=dict(t_j=t_j, Delta=Delta, mu=mu ,mu_sc=mu_sc, phi=phi)

bands = kwant.physics.Bands(syst.leads[0], params=params)
energies = [bands(k) for k in momenta]
```

```python
plt.figure()
plt.plot(np.array(momenta)/3/pi, np.array(energies)/Delta,'b')
# plt.plot((momenta), (energies))
plt.ylim(-2,2)
plt.xlabel("momentum [(lattice constant)^-1]")
plt.ylabel(r"energy [$\Delta$]")
plt.grid()
plt.show()
```
